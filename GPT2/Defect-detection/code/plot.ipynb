```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("trainlog.csv")
x = list(df['step'])
y = list(df['trainloss'])
d = []
for i in range(len(x)):
    d.append([x[i],y[i]])
df2 = pd.DataFrame(columns=['train step','trainloss'],data=d)
```

```python
sns.set_theme(style='darkgrid')
plt.figure()
fig, ax =plt.subplots(1,2,constrained_layout=True, figsize=(6, 3))
```

```python
f = sns.lineplot(x='train step',y='trainloss',data=df2,ax = ax[0])
f.set_title("trainloss change")
```

```python
validloss = df['validloss']
acc = df['acc']
f1socre = df['f1-score']
```

```python
rx = []
for i in range(len(x)):
    if not pd.isna(validloss[i]):
        rx.append(i)
```

```python
validloss =validloss.dropna(axis=0,how='all')
acc = acc.dropna(axis=0,how='all')
```

```python
d = []
for i in range(len(rx)):
    d.append([validloss[rx[i]],acc[rx[i]]])
```

```python
df3 = pd.DataFrame(columns=['validloss','acc'],data=d,index= rx)
```

```python
f = sns.lineplot(data=df3,ax = ax[1])
f.set_xlabel("train step")
f.set_title("evaluate metircs change")
```
